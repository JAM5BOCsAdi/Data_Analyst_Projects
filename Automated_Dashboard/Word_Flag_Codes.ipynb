```python
import requests
from bs4 import BeautifulSoup

# Correctly define the URL
url = "https://www.worldatlas.com/countries"

# Properly format the headers as a dictionary
my_headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/128.0.0.0 Safari/537.36"}

# Send a GET request with headers
page = requests.get(url, headers=my_headers)

# Parse the page content with BeautifulSoup
soup = BeautifulSoup(page.content, 'html.parser')

# You can now work with the `soup` object
print(soup.prettify())  # To visualize the HTML structure
```

```python
item_iso3 = soup.find_all('td',id='item_iso3')
print (item_iso3)
```

```python
# Initialize an empty array to store the items
iso3_codes = []

# Iterate over the elements to extract text and save it in the array
for item in item_iso3:
    iso3_codes.append(item.get_text())

# Print the array to verify the contents
print(iso3_codes)
```
